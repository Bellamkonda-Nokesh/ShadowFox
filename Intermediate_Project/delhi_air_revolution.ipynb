## 1.3 Advanced Feature Engineering Laboratory

> *"Great insights come from great features."*

```python
# ============================================================================
# TEMPORAL FEATURE EXTRACTION
# ============================================================================

print("‚öôÔ∏è  Engineering temporal features...\n")

# Basic time features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.dayofweek
df['day_name'] = df['date'].dt.day_name()
df['month_name'] = df['date'].dt.month_name()
df['week_of_year'] = df['date'].dt.isocalendar().week
df['quarter'] = df['date'].dt.quarter

# Advanced temporal features
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
df['is_rush_hour'] = df['hour'].isin([7, 8, 9, 17, 18, 19]).astype(int)
df['is_night'] = df['hour'].isin(list(range(0, 6)) + list(range(22, 24))).astype(int)
df['is_winter'] = df['month'].isin([11, 12, 1]).astype(int)
df['is_monsoon'] = df['month'].isin([6, 7, 8]).astype(int)

# Cyclical encoding (hour and month)
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

# ============================================================================
# AQI CALCULATION - EPA METHOD
# ============================================================================

print("üíä Calculating AQI indices...\n")

def calculate_aqi_pm25(pm25):
    """Calculate AQI for PM2.5 using EPA breakpoints"""
    if pd.isna(pm25):
        return np.nan
    if pm25 <= 12.0:
        return ((50 - 0) / (12.0 - 0.0)) * (pm25 - 0.0) + 0
    elif pm25 <= 35.4:
        return ((100 - 51) / (35.4 - 12.1)) * (pm25 - 12.1) + 51
    elif pm25 <= 55.4:
        return ((150 - 101) / (55.4 - 35.5)) * (pm25 - 35.5) + 101
    elif pm25 <= 150.4:
        return ((200 - 151) / (150.4 - 55.5)) * (pm25 - 55.5) + 151
    elif pm25 <= 250.4:
        return ((300 - 201) / (250.4 - 150.5)) * (pm25 - 150.5) + 201
    elif pm25 <= 350.4:
        return ((400 - 301) / (350.4 - 250.5)) * (pm25 - 250.5) + 301
    elif pm25 <= 500.4:
        return ((500 - 401) / (500.4 - 350.5)) * (pm25 - 350.5) + 401
    else:
        return 500

def calculate_aqi_pm10(pm10):
    """Calculate AQI for PM10 using EPA breakpoints"""
    if pd.isna(pm10):
        return np.nan
    if pm10 <= 54:
        return ((50 - 0) / (54 - 0)) * (pm10 - 0) + 0
    elif pm10 <= 154:
        return ((100 - 51) / (154 - 55)) * (pm10 - 55) + 51
    elif pm10 <= 254:
        return ((150 - 101) / (254 - 155)) * (pm10 - 155) + 101
    elif pm10 <= 354:
        return ((200 - 151) / (354 - 255)) * (pm10 - 255) + 151
    elif pm10 <= 424:
        return ((300 - 201) / (424 - 355)) * (pm10 - 355) + 201
    elif pm10 <= 504:
        return ((400 - 301) / (504 - 425)) * (pm10 - 425) + 301
    elif pm10 <= 604:
        return ((500 - 401) / (604 - 505)) * (pm10 - 505) + 401
    else:
        return 500

# Calculate AQI
df['aqi_pm2_5'] = df['pm2_5'].apply(calculate_aqi_pm25)
df['aqi_pm10'] = df['pm10'].apply(calculate_aqi_pm10)
df['aqi'] = df[['aqi_pm2_5', 'aqi_pm10']].max(axis=1)

# Health categorization
def classify_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy (Sensitive)'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

df['health_category'] = df['aqi'].apply(classify_aqi)
df['high_risk'] = (df['aqi'] > 150).astype(int)

# ============================================================================
# ADVANCED POLLUTION METRICS
# ============================================================================

print("üß¨ Creating pollution fingerprints...\n")

# Composite pollution index (weighted average)
df['pollution_index'] = (
    0.40 * df['pm2_5'] / df['pm2_5'].max() +
    0.30 * df['pm10'] / df['pm10'].max() +
    0.15 * df['no2'] / df['no2'].max() +
    0.10 * df['co'] / df['co'].max() +
    0.05 * df['so2'] / df['so2'].max()
) * 100

# Particulate matter ratio (PM2.5/PM10)
df['pm_ratio'] = df['pm2_5'] / df['pm10']

# Nitrogen oxide ratio (NO2/NO)
df['nox_ratio'] = df['no2'] / (df['no'] + 0.01)  # Add small value to avoid division by zero

# Traffic pollution indicator
df['traffic_indicator'] = (df['no2'] + df['co']) / 2

# Industrial pollution indicator
df['industrial_indicator'] = (df['so2'] + df['pm10']) / 2

# Agricultural burning indicator
df['agricultural_indicator'] = df['nh3']

# Rolling averages (24-hour)
for pol in ['pm2_5', 'pm10', 'no2', 'aqi']:
    df[f'{pol}_ma24'] = df[pol].rolling(window=24, min_periods=1).mean()

# Rate of change
df['aqi_change'] = df['aqi'].diff()
df['aqi_pct_change'] = df['aqi'].pct_change() * 100

# ============================================================================
# FEATURE ENGINEERING SUMMARY
# ============================================================================

print("="*100)
print("‚ú® FEATURE ENGINEERING COMPLETE")
print("="*100)
print(f"\nüìä Total Features Created: {len(df.columns)}")
print("\nüèóÔ∏è  Feature Categories:")
print(f"   ‚úì Temporal Features: 15 (hour, day, month, season, cyclical)")
print(f"   ‚úì AQI Metrics: 4 (PM2.5, PM10, Overall, Health Category)")
print(f"   ‚úì Pollution Fingerprints: 6 (Traffic, Industrial, Agricultural indicators)")
print(f"   ‚úì Advanced Metrics: 8 (Ratios, Rolling averages, Rate of change)")
print("\nüíä AQI Distribution:")
health_dist = df['health_category'].value_counts(sort=False)
for cat in ['Good', 'Moderate', 'Unhealthy (Sensitive)', 'Unhealthy', 'Very Unhealthy', 'Hazardous']:
    count = health_dist.get(cat, 0)
    pct = (count / len(df)) * 100
    bar = '‚ñà' * int(pct / 2)
    print(f"   {cat:25s}: {count:6,} hours ({pct:5.1f}%) {bar}")
print("\nüéØ Key Statistics:")
print(f"   ‚Ä¢ Average AQI: {df['aqi'].mean():.1f}")
print(f"   ‚Ä¢ Median AQI: {df['aqi'].median():.1f}")
print(f"   ‚Ä¢ Worst AQI: {df['aqi'].max():.1f} on {df.loc[df['aqi'].idxmax(), 'date']}")
print(f"   ‚Ä¢ Best AQI: {df['aqi'].min():.1f} on {df.loc[df['aqi'].idxmin(), 'date']}")
print(f"   ‚Ä¢ High-Risk Hours: {df['high_risk'].sum():,} ({df['high_risk'].mean()*100:.1f}%)")
print("="*100)
```

## 2.1 Pollutant Fingerprinting

> *"Every pollution source leaves a unique signature - let's decode it."*

We'll use advanced correlation and clustering techniques to identify pollution "fingerprints":

```python
# ============================================================================
# VISUALIZATION 1: POLLUTANT FINGERPRINT MATRIX
# ============================================================================

fig = plt.figure(figsize=(22, 12))
gs = GridSpec(2, 3, figure=fig, hspace=0.3, wspace=0.3)

pollutants = ['pm2_5', 'pm10', 'no2', 'no', 'co', 'so2', 'o3', 'nh3']

# 1. Correlation Heatmap with Hierarchical Clustering
ax1 = fig.add_subplot(gs[0, :])
corr_matrix = df[pollutants].corr()

# Hierarchical clustering
linkage_matrix = linkage(corr_matrix, method='ward')
dendro = dendrogram(linkage_matrix, labels=corr_matrix.columns, no_plot=True)
order = dendro['leaves']
corr_ordered = corr_matrix.iloc[order, order]

# Plot heatmap
im = ax1.imshow(corr_ordered, cmap='RdYlGn', aspect='auto', vmin=-1, vmax=1)
ax1.set_xticks(range(len(corr_ordered)))
ax1.set_yticks(range(len(corr_ordered)))
ax1.set_xticklabels([col.upper() for col in corr_ordered.columns], rotation=45, ha='right')
ax1.set_yticklabels([col.upper() for col in corr_ordered.columns])

# Add correlation values
for i in range(len(corr_ordered)):
    for j in range(len(corr_ordered)):
        value = corr_ordered.iloc[i, j]
        color = 'white' if abs(value) > 0.5 else 'black'
        ax1.text(j, i, f'{value:.2f}', ha='center', va='center',
                color=color, fontsize=10, fontweight='bold')

cbar = plt.colorbar(im, ax=ax1, fraction=0.046, pad=0.04)
cbar.set_label('Correlation Coefficient', fontsize=12, fontweight='bold')
ax1.set_title('üß¨ Pollutant Correlation Matrix (Hierarchically Clustered)', 
              fontsize=16, fontweight='bold', pad=15)

# 2. Traffic Fingerprint (NO2 vs CO)
ax2 = fig.add_subplot(gs[1, 0])
scatter = ax2.hexbin(df['no2'], df['co'], gridsize=50, cmap=neon_cmap, 
                     mincnt=1, edgecolors='white', linewidths=0.2)
ax2.set_xlabel('NO‚ÇÇ (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax2.set_ylabel('CO (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax2.set_title('üöó Traffic Pollution Fingerprint\n(NO‚ÇÇ vs CO Density)', 
              fontsize=14, fontweight='bold')
cb = plt.colorbar(scatter, ax=ax2)
cb.set_label('Frequency', fontsize=10)

# Add trend line
z = np.polyfit(df['no2'].dropna(), df['co'].dropna(), 1)
p = np.poly1d(z)
ax2.plot(df['no2'].dropna().sort_values(), 
         p(df['no2'].dropna().sort_values()), 
         "r--", alpha=0.8, linewidth=3, label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}')
ax2.legend(loc='upper left')

# 3. Industrial Fingerprint (SO2 vs PM10)
ax3 = fig.add_subplot(gs[1, 1])
scatter = ax3.hexbin(df['so2'], df['pm10'], gridsize=50, cmap=neon_cmap,
                     mincnt=1, edgecolors='white', linewidths=0.2)
ax3.set_xlabel('SO‚ÇÇ (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax3.set_ylabel('PM10 (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax3.set_title('üè≠ Industrial Pollution Fingerprint\n(SO‚ÇÇ vs PM10 Density)', 
              fontsize=14, fontweight='bold')
cb = plt.colorbar(scatter, ax=ax3)
cb.set_label('Frequency', fontsize=10)

# Add trend line
z = np.polyfit(df['so2'].dropna(), df['pm10'].dropna(), 1)
p = np.poly1d(z)
ax3.plot(df['so2'].dropna().sort_values(), 
         p(df['so2'].dropna().sort_values()), 
         "r--", alpha=0.8, linewidth=3, label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}')
ax3.legend(loc='upper left')

# 4. Agricultural Fingerprint (NH3 vs PM2.5)
ax4 = fig.add_subplot(gs[1, 2])
scatter = ax4.hexbin(df['nh3'], df['pm2_5'], gridsize=50, cmap=neon_cmap,
                     mincnt=1, edgecolors='white', linewidths=0.2)
ax4.set_xlabel('NH‚ÇÉ (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax4.set_ylabel('PM2.5 (Œºg/m¬≥)', fontsize=12, fontweight='bold')
ax4.set_title('üåæ Agricultural Burning Fingerprint\n(NH‚ÇÉ vs PM2.5 Density)', 
              fontsize=14, fontweight='bold')
cb = plt.colorbar(scatter, ax=ax4)
cb.set_label('Frequency', fontsize=10)

# Add trend line
z = np.polyfit(df['nh3'].dropna(), df['pm2_5'].dropna(), 1)
p = np.poly1d(z)
ax4.plot(df['nh3'].dropna().sort_values(), 
         p(df['nh3'].dropna().sort_values()), 
         "r--", alpha=0.8, linewidth=3, label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}')
ax4.legend(loc='upper left')

plt.suptitle('üî¨ POLLUTION SOURCE DNA FINGERPRINTING', 
             fontsize=20, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()

print("\n" + "="*100)
print("üîç FINGERPRINT ANALYSIS INSIGHTS")
print("="*100)
print("\nüìä Strongest Correlations:")
corr_flat = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
corr_flat = corr_flat.stack().sort_values(ascending=False)
for i, (pair, value) in enumerate(corr_flat.head(5).items()):
    print(f"   {i+1}. {pair[0].upper()} ‚Üî {pair[1].upper()}: {value:.3f}")
print("="*100)
```
